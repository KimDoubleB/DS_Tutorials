```python
import pygal
from pygal.maps.world import World
from pygal.style import Style
custom_style = Style(colors=('#FFFF00', '#00FFFF', '#FF00FF'))



def world_map_plot(df, result, filename):
    #worldmap_chart = World()
    worldmap_chart = pygal.maps.world.World(style=custom_style)
    # make space data (for dynamically allocation)
    c = []
    for i in range(max(result) + 1):
        c.append([])
    
    # add data to list of cluster number
    for i in range(len(result)):
        c[result[i]].append((df.iloc[i,2]).lower())

    # add list to worldmap_chart
    for i, v in enumerate(c):
        worldmap_chart.add('Class' + str(i), v)
        
    worldmap_chart.render_to_file(filename + '.html')
 
```

```python
def purity_calculator(cluster_data, class_data):
    # data [class, cluster]
    df_p = pd.DataFrame({'class':class_data.tolist(), 'cluster':cluster_data})
    equal_max_class = 0

    # get purity value
    for i in pd.unique(df_p.cluster):
        if i== -1:
            continue
        temp = df_p[df_p['cluster'] == i]
        class_of_cluster = temp.mode()['class'][0]
        print('cluster {} - class {}' .format(i, class_of_cluster))
        equal_max_class += len(temp[temp['class']==class_of_cluster])
    purity = equal_max_class / len(cluster_data)
    return purity

```

```python
from sklearn.mixture import GaussianMixture
while(1):
    model = GaussianMixture(n_components=3, init_params='random')
    model.fit(X)
    # predict latent values
    em_result = model.predict(X)
    em_purity_result = purity_calculator(em_result, df_class)
    if em_purity_result > 0.52:
        break
```

```python
world_map_plot(df2, em_result, 'em_1971')
print("EM purity:")
print(em_purity_result)
```

```python
from sklearn.cluster import DBSCAN

dbscan_purity_result = 0
for i in range(1, 15):
    dbscan = DBSCAN(eps=i, min_samples=2).fit(X)
    temp_result = purity_calculator(dbscan.labels_, df_class)
    if temp_result > dbscan_purity_result:
        dbscan_result = dbscan.labels_
        dbscan_purity_result = temp_result
```
